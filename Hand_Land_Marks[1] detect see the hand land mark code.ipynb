```python
import cv2
import time 
import mediapipe
```

```python
# result=cv2.VideoWriter('handlandmarks.avi', cv2.VideoWriter_fourcc(*"XVID"),30,(640,480))
result = cv2.VideoWriter('handlandmarks.mp4', cv2.VideoWriter_fourcc(*'H264'), 15, (640, 480))

cap=cv2.VideoCapture(0) 


mphands=mediapipe.solutions.hands
hands = mphands.Hands(
    static_image_mode=False, 
    max_num_hands=5)
mpDraw=mediapipe.solutions.drawing_utils

pTime=0
cTime=0

while True:
    _,img=cap.read()
    imgRGB= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results=hands.process(imgRGB)
    print(results)
    print(results.multi_hand_landmarks)
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:

            mpDraw.draw_landmarks(img,handLms,mphands.HAND_CONNECTIONS)

    cTime=time.time()   
    fps=1/(cTime-pTime)
    pTime=cTime
    
    cv2.putText(img,str(int(fps)),(15,60),cv2.FONT_HERSHEY_PLAIN,3,(0,0,255))
    result.write(img)
    cv2.imshow('Camera',img)
    key=cv2.waitKey(1)
    if key==ord('q'):
      break

cv2.destroyAllWindows()
cap.release()
```

```python
# result=cv2.VideoWriter('handlandmarks.avi', cv2.VideoWriter_fourcc(*"XVID"),30,(640,480))
result = cv2.VideoWriter('handlandmarks.mp4', cv2.VideoWriter_fourcc(*'H264'), 30, (640, 480))

cap=cv2.VideoCapture(0)


mphands=mediapipe.solutions.hands
hands=mphands.Hands(static_image_mode=False,max_num_hands=5)
mpDraw=mediapipe.solutions.drawing_utils

pTime=0
cTime=0

while True:
    _,img=cap.read()
    imgRGB= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results=hands.process(imgRGB)
    print(results)
    print(results.multi_hand_landmarks)
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for idd, Lm in enumerate(handLms.landmark):
                print(idd,Lm)
                h,w,c=img.shape
                cx , cy = int(Lm.x*w)  , int(Lm.y*h)
                print(f"Landmark {idd}: Normalized({Lm.x:.2f}, {Lm.y:.2f}) → Pixels({cx}, {cy})")
                print(idd,cx,cy)
                if idd==8: #change number 8 for change circle position
                    cv2.circle(img,(cx,cy),11,(255,0,255),cv2.FILLED)
                if idd==12:
                    cv2.circle(img,(cx,cy),11,(255,0,255),cv2.FILLED)      
            mpDraw.draw_landmarks(img,handLms,mphands.HAND_CONNECTIONS)


    cTime=time.time()   
    fps=1/(cTime-pTime)
    pTime=cTime
    
    cv2.putText(img,str(int(fps)),(15,60),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0))
    result.write(img)
    cv2.imshow('fram',img)
    key=cv2.waitKey(1)
    if key==ord('q'):
      break

cv2.destroyAllWindows()
cap.release()
```
